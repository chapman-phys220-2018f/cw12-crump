# The Assigment

In this assignment, we considered the following parametrized system of differential equations:
$$ x'(t) = y(t) $$
$$ my'(t) = -\nu y(t) + x(t) - x'(t)^3 + F\cos(\omega t) $$
With conditions, $m = 1, \nu = 0.25, \omega = 1$, and then changed only $F, x(0),$ and $y(0)$

We plotted the approximated solution line as a function of $t$ in the first plot, then plotted the phase portrait in the second plot, and finally plotted the Poincare section in the last plot for each of the following examples.

```python
sb.rungekutta4th(0,2*np.pi,50,-0.9,0.0,0.18)
```

With these values, the system is very well behaved (and even gives us a lovely avacado as a phase portrait.)

Things only change slightly when we invert to $x = 0.9$

```python
sb.rungekutta4th(0,2*np.pi,50,0.9,0.0,0.18)
```

Again the plot is well behaved, but now we have a flipped avacado!

Things start getting interesting when we change the driving force, in the next we increase it to $F = 0.25$

```python
sb.rungekutta4th(0,2*np.pi,50,0.2,0.1,0.25)
```

Now the plot looks very different, and we can see from the time plot that it has an initial chaotic type behavior and then settles down into a figure pretty similar to the last two. We see a nice difference in the phase portrait, though, we got a mutant avacado! It looks like the system has two semi-equilibria at the pits of the avacados.

Lets see what happens when we change $F = 0.4$

```python
sb.rungekutta4th(0,2*np.pi,50,0,0.0,0.4)
```

The system is now real chaotic, it looks like it never settles down into any particular motion at any point in time (take that, small angle approximation!) The phase portrait got way more chaotic too and we lost our avacados (crying emoji) However, what we lost in produce we gained in the how much more we see from the Poincare section, which has a more filled out shape in the next graph.

```python
sb.rungekutta4th(0,2*np.pi,1000,0,0.0,0.4,plotall = False)
```

The structure is pretty interesting! It looks like the figure is wrapped around the avacado pits and is very smooth. I have no idea how to analyze this, but I can say that I sure do like it.
